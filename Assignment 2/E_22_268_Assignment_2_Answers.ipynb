#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



## Answer



### 1. Probability Space
Let the sample space be
$$\Omega = \{0, 1, 2, 3\}$$
where each outcome represents the number of emergency transformer failures in a single day.

Let the sigma-algebra be the power set of $\Omega$:
$$\mathcal{F} = 2^{\Omega}$$

Define the probability measure $P$ by:
$$P(0)=0.50, \quad P(1)=0.30, \quad P(2)=0.15, \quad P(3)=0.05$$

For any event $A \subseteq \Omega$:
$$P(A) = \sum_{\omega \in A} P(\omega)$$

Thus, $(\Omega, \mathcal{F}, P)$ is a valid probability space.

### 2. Random Variable
Define the random variable $X : \Omega \to \mathbb{R}$ by:
$$X(\omega) = \omega, \quad \omega \in \Omega$$
This variable represents the number of transformer failures observed in a day.

### 3. Probability Mass Function
The probability mass function (PMF) of $X$ is:
$$p_X(x) =
\begin{cases}
0.50, & x = 0 \\
0.30, & x = 1 \\
0.15, & x = 2 \\
0.05, & x = 3 \\
0, & \text{otherwise}
\end{cases}$$

### 4. Expectation
The expected value of $X$ is:
$$E[X] = \sum_x x p_X(x)$$

Therefore:
$$E[X] = 0(0.50) + 1(0.30) + 2(0.15) + 3(0.05) = 0.75$$

### 5. Variance
First, compute the second moment:
$$E[X^2] = 0^2(0.50) + 1^2(0.30) + 2^2(0.15) + 3^2(0.05) = 1.35$$

Then:
$$\operatorname{Var}(X) = E[X^2] - (E[X])^2 = 1.35 - (0.75)^2 = 0.7875$$

### 6. Required Probabilities
*   **At least one failure:**
    $$P(X \ge 1) = P(X=1) + P(X=2) + P(X=3) = 0.50$$
*   **At least two failures:**
    $$P(X \ge 2) = P(X=2) + P(X=3) = 0.20$$

### 7. Interpretation
The expectation $E[X] = 0.75$ represents the average number of emergency transformer failures per day over a long period. The variance $0.7875$ indicates the variability in the daily number of failures, showing how much the count typically fluctuates around that average.


# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


##  Answer



### 1. Probability Space
Since the proportion of defective items can take any value between 0 and 1, we choose the sample space as:
$$\Omega = [0, 1]$$

Let $\mathcal{F}$ be the Borel $\sigma$-algebra on $[0, 1]$, denoted by:
$$\mathcal{F} = \mathcal{B}([0, 1])$$

Define a probability measure $P$ on $(\Omega, \mathcal{F})$ by:
$$P(A) = \int_A f(x) \, dx, \quad A \in \mathcal{F}$$
where the density function $f$ is defined below. Thus, $(\Omega, \mathcal{F}, P)$ forms a valid probability space.

### 2. Definition of the Random Variable
Define the random variable $X : \Omega \to \mathbb{R}$ by:
$$X(\omega) = \omega, \quad \omega \in [0, 1]$$
Here, $X$ represents the proportion of defective items produced in a batch on a given day.

### 3. Probability Density Function
The random variable $X$ follows a Beta distribution with parameters $\alpha = 2$ and $\beta = 18$. Its probability density function (PDF) is:
$$f_X(x) =
\begin{cases}
\dfrac{1}{B(2,18)} x(1-x)^{17}, & 0 < x < 1 \\
0, & \text{otherwise}
\end{cases}$$

### 4. Verification of a Valid Density
First, observe that $f_X(x) \ge 0$ for all $x \in \mathbb{R}$. Next, we integrate over the real line:
$$\int_{-\infty}^{\infty} f_X(x) \, dx = \int_0^1 \frac{1}{B(2,18)} x(1-x)^{17} \, dx = \frac{1}{B(2,18)} B(2,18) = 1$$
Hence, $f_X(x)$ is a valid probability density function.

### 5. Expectation
For a $\text{Beta}(\alpha, \beta)$ random variable, the expected value is:
$$E[X] = \frac{\alpha}{\alpha + \beta}$$
Substituting $\alpha = 2$ and $\beta = 18$:
$$E[X] = \frac{2}{20} = 0.10$$

### 6. Variance
The variance of a $\text{Beta}(\alpha, \beta)$ distribution is:
$$\operatorname{Var}(X) = \frac{\alpha \beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$$
Substituting our values:
$$\operatorname{Var}(X) = \frac{2 \cdot 18}{20^2 \cdot 21} = \frac{36}{8400} \approx 0.00429$$

### 7. Probability: Defective Proportion is at Most 0.10
This is calculated as the area under the curve from 0 to 0.10:
$$P(X \le 0.10) = \int_0^{0.10} \frac{1}{B(2,18)} x(1-x)^{17} \, dx$$
This can be expressed using the regularized incomplete Beta function:
$$P(X \le 0.10) = I_{0.10}(2, 18)$$

### 8. Probability: Defective Proportion Between 0.05 and 0.15
$$P(0.05 \le X \le 0.15) = \int_{0.05}^{0.15} \frac{1}{B(2,18)} x(1-x)^{17} \, dx$$
Equivalently:
$$P(0.05 \le X \le 0.15) = I_{0.15}(2, 18) - I_{0.05}(2, 18)$$



### 9. Why the Beta Distribution is Appropriate
The Beta distribution is ideal for modeling proportions because it is strictly defined on the interval $[0, 1]$. It is flexible enough to capture various shapes, such as skewness toward low values (which we expect in quality control), making it perfect for representing defective fractions in manufacturing.

### 10. Interpretation
*   **Expectation ($E[X] = 0.10$):** This indicates that, on average, 10% of the items produced in a batch are defective.
*   **Variance ($\approx 0.00429$):** This measures how much the defective proportion fluctuates from day to day. A low variance suggests high consistency in the manufacturing process.


### The plot of the $\beta$-distribution

```python

import numpy as np
import plotly.graph_objects as go
from scipy.stats import beta

def plot_beta_distribution(alpha, beta_val):
    """
    Plots the Beta distribution PDF for given alpha and beta parameters using Plotly.
    """
    # Create a range of x values from 0 to 1
    x = np.linspace(0, 1, 1000)

    # Calculate the PDF values for the Beta distribution
    y = beta.pdf(x, alpha, beta_val)

    # Create the Plotly figure
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=x,
        y=y,
        mode='lines',
        name=f'Beta($\\alpha={alpha}, \\beta={beta_val}$)',
        line=dict(width=2)
    ))

    # Update layout for better readability
    fig.update_layout(
        title=f'Beta Distribution Density Function ($\\alpha={alpha}, \\beta={beta_val}$)',
        xaxis_title='x',
        yaxis_title='PDF(x)',
        template='plotly_white'
    )

    # To display the plot, use fig.show()
    return fig

# Example usage:
fig = plot_beta_distribution(alpha=2, beta_val=18)
fig.show()
```

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


##  Answer



### 1. Support of the Random Variable
The random variable $X$ represents the time (in hours) required to install the backup generator. Since the installation cannot take less than 4 hours or more than 10 hours, the support of $X$ is:
$$4 \le X \le 10$$

### 2. Probability Density Function
The installation time follows a **Triangular distribution** with parameters $a = 4$ (minimum), $b = 10$ (maximum), and $c = 6$ (most likely). The general PDF is:

$$f_X(x) =
\begin{cases}
\dfrac{2(x-a)}{(b-a)(c-a)}, & a \le x < c \\[6pt]
\dfrac{2}{b-a}, & x = c \\[6pt]
\dfrac{2(b-x)}{(b-a)(b-c)}, & c < x \le b \\[6pt]
0, & \text{otherwise}
\end{cases}$$

Substituting our values ($a=4, b=10, c=6$):
$$f_X(x) =
\begin{cases}
\dfrac{x-4}{6}, & 4 \le x < 6 \\[6pt]
\dfrac{1}{3}, & x = 6 \\[6pt]
\dfrac{10-x}{12}, & 6 < x \le 10 \\[6pt]
0, & \text{otherwise}
\end{cases}$$

### 3. Verification of a Valid Density
To confirm this is a valid PDF, we integrate over the support:
$$\int_4^6 \frac{x-4}{6} \, dx + \int_6^{10} \frac{10-x}{12} \, dx$$

Evaluating the parts:
*   $\int_4^6 \frac{x-4}{6} \, dx = \frac{1}{6} \left[ \frac{(x-4)^2}{2} \right]_4^6 = \frac{1}{3}$
*   $\int_6^{10} \frac{10-x}{12} \, dx = \frac{1}{12} \left[ -\frac{(10-x)^2}{2} \right]_6^{10} = \frac{2}{3}$

Summing them: $\frac{1}{3} + \frac{2}{3} = 1$. This confirms the density is valid.

### 4. Expectation
The expected value for a Triangular distribution is:
$$E[X] = \frac{a + b + c}{3}$$
$$E[X] = \frac{4 + 10 + 6}{3} = \frac{20}{3} \approx 6.67 \text{ hours}$$

### 5. Variance
The variance is calculated as:
$$\operatorname{Var}(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}$$
$$\operatorname{Var}(X) = \frac{16 + 100 + 36 - 40 - 24 - 60}{18} = \frac{28}{18} = \frac{14}{9} \approx 1.56$$

### 6. Interpretation of Parameters
*   **$a = 4$:** The "best-case" scenario (fastest possible time).
*   **$b = 10$:** The "worst-case" scenario (slowest possible time).
*   **$c = 6$:** The most likely duration based on typical conditions.

### 7. Justification for the Triangular Model
The Triangular distribution is highly effective when historical data is scarce but expert estimates provide bounds and a "most likely" value. It constrains the variable to a realistic range while accounting for skewness.

### 8. Probability: Installation at Most 5 Hours
Since $5 < c$, we use the left side of the density function:
$$P(X \le 5) = \int_4^5 \frac{x-4}{6} \, dx = \left[ \frac{(x-4)^2}{12} \right]_4^5 = \frac{1}{12} \approx 0.0833$$

### 9. Probability: Installation Between 5 and 8 Hours
We split the integral at the peak ($x=6$):
$$P(5 \le X \le 8) = \int_5^6 \frac{x-4}{6} \, dx + \int_6^8 \frac{10-x}{12} \, dx$$
$$P(5 \le X \le 8) = \frac{1}{4} + \frac{1}{2} = \frac{3}{4} = 0.75$$

### 10. Use in Project Planning
This model assists managers in **risk assessment**. By knowing there is a 75% chance the installation takes between 5 and 8 hours, they can better allocate staffing and set more realistic project milestones.


### Plot of the triangular distribution

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import triang

# Parameters given
a = 4
b = 10
c_mode = 6

# Scipy parametrization
# loc is the minimum (a)
# scale is the width (b - a)
# c_scipy is the ratio of the mode relative to the width (c - a) / (b - a)
loc = a
scale = b - a
c_scipy = (c_mode - a) / (b - a)

# Generate data points
x = np.linspace(0, b, 500)
y = triang.pdf(x, c=c_scipy, loc=loc, scale=scale)

# Create the Plotly figure
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode='lines',
    fill='tozeroy',
    name='Triangular Distribution',
    line=dict(color='royalblue', width=2)
))

# Layout updates
fig.update_layout(
    title=f'Triangular Distribution (a={a}, b={b}, c={c_mode})',
    xaxis_title='Time (hours)',
    yaxis_title='Probability Density',
    template='plotly_white'
)

# Display the plot
fig.show()
```

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


## Answer


The task duration $X$ (in days) is modeled using a **PERT distribution** with the following parameters:
*   **Minimum ($a$):** 8
*   **Maximum ($b$):** 20
*   **Most Likely ($c$):** 11

The PERT distribution is based on a Beta distribution scaled to the interval $[a, b]$, with shape parameters defined as:
$$\alpha = 1 + \frac{4(c-a)}{b-a}, \qquad \beta = 1 + \frac{4(b-c)}{b-a}$$

### 1. Shape Parameters ($\alpha$ and $\beta$)
Substituting our values:
$$\alpha = 1 + \frac{4(11-8)}{20-8} = 1 + \frac{12}{12} = 2$$
$$\beta = 1 + \frac{4(20-11)}{20-8} = 1 + \frac{36}{12} = 4$$

Thus, the internal Beta parameters are **$\alpha = 2$** and **$\beta = 4$**.

### 2. Support
The support of the random variable $X$ is the interval where completion is possible:
$$8 \le X \le 20$$
### 3.Probability Density Function
The probability density function (PDF) is:
$$f_X(x) = \frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \quad a \le x \le b$$

Substituting $a=8, b=20, \alpha=2, \beta=4$:
$$f_X(x) = \frac{(x-8)(20-x)^3}{B(2,4) \cdot 12^5}$$

Using the identity $B(2,4) = \frac{\Gamma(2)\Gamma(4)}{\Gamma(6)} = \frac{1! \cdot 3!}{5!} = \frac{6}{120} = \frac{1}{20}$, the density simplifies to:
$$f_X(x) = \frac{20(x-8)(20-x)^3}{12^5}, \quad 8 \le x \le 20$$

### 4. Probability Space
*   **Sample Space:** $\Omega = [8, 20]$
*   **Sigma-algebra:** $\mathcal{F} = \mathcal{B}([8, 20])$ (Borel $\sigma$-algebra)
*   **Probability Measure:** $P(A) = \int_A f_X(x) \, dx$
### 5.Random Variable
 $X(\omega) = \omega$, representing the task duration in days.

### 6. Expected Value
For a PERT distribution, the expected value is a weighted average:
$$E[X] = \frac{a + 4c + b}{6}$$
$$E[X] = \frac{8 + 4(11) + 20}{6} = \frac{72}{6} = 12 \text{ days}$$

### 7. Variance
Using the common approximation for PERT variance:
$$\operatorname{Var}(X) = \frac{(E[X]-a)(b-E[X])}{7}$$
$$\operatorname{Var}(X) = \frac{(12-8)(20-12)}{7} = \frac{32}{7} \approx 4.57$$

### 8.The standard deviation
$$\sigma = \sqrt{4.57} \approx 2.14$$


### 9. Comparison: PERT vs. Triangular Distribution
The PERT distribution is often preferred over the Triangular distribution because it:
*   **Weights the mode more heavily:** The "4" in the $4c$ term gives the most likely value more influence.
*   **Is smoother:** It lacks the "peak" or "kink" found at the mode of a Triangular distribution.
*   **Reduces variance:** It typically produces a narrower distribution, reflecting more confidence in the most likely estimate.

### 10. Practical Interpretation
*   **Optimistic ($a=8$):** The best-case scenario.
*   **Pessimistic ($b=20$):** The worst-case scenario.
*   **Most Likely ($c=11$):** The "normal" expectation.
*   **Expected ($E[X]=12$):** The value used for the master project schedule.
*   **Uncertainty ($\sigma \approx 2.14$):** Helps planners decide how much "buffer" time to add to the project to account for risks.


### Plot of the triangle and PERT distributions

```python

import numpy as np
import plotly.graph_objects as go
from scipy.stats import triang, beta

# Parameters
a = 4
b = 10
c_mode = 6

# --- Triangular Distribution Calculation ---
# Scipy parametrization for Triangular
c_scipy = (c_mode - a) / (b - a)
loc = a
scale = b - a

# --- PERT Distribution Calculation ---
# PERT is a Beta distribution with specific alpha and beta parameters
# alpha = 1 + 4 * (c - a) / (b - a)
# beta = 1 + 4 * (b - c) / (b - a)
alpha_pert = 1 + 4 * (c_mode - a) / (b - a)
beta_pert = 1 + 4 * (b - c_mode) / (b - a)

# Generate data points
x = np.linspace(0, b, 500)
y_triang = triang.pdf(x, c=c_scipy, loc=loc, scale=scale)
y_pert = beta.pdf(x, alpha_pert, beta_pert, loc=a, scale=b-a)

# Create the Plotly figure
fig = go.Figure()

# Add Triangular Distribution
fig.add_trace(go.Scatter(
    x=x,
    y=y_triang,
    mode='lines',
    name='Triangular Distribution',
    line=dict(color='royalblue', width=2, dash='dash')
))

# Add PERT Distribution
fig.add_trace(go.Scatter(
    x=x,
    y=y_pert,
    mode='lines',
    name='PERT Distribution',
    line=dict(color='firebrick', width=2)
))

# Layout updates
fig.update_layout(
    title=f'Comparison: Triangular vs PERT (a={a}, b={b}, c={c_mode})',
    xaxis_title='Time (hours)',
    yaxis_title='Probability Density',
    template='plotly_white'
)

# Display the plot
fig.show()
```

#Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

## Answer


### 1. Determining the Maximum Time ($b$)
In a PERT distribution, the expected value of the task duration is given by the weighted average:
$$E[X] = \frac{a + 4c + b}{6}$$

We are given:
*   Minimum time ($a$) = 5 days
*   Most likely time ($c$) = 11 days

The problem states that the expected duration must coincide with the most likely time ($E[X] = c$). We can set up the equation as follows:
$$\frac{a + 4c + b}{6} = c$$

Multiplying both sides by 6:
$$a + 4c + b = 6c$$

Rearranging to solve for $b$:
$$b = 6c - 4c - a$$
$$b = 2c - a$$

Substituting our values:
$$b = 2(11) - 5 = 17$$

**Result:** The maximum possible task duration is **17 days**.


### 2. Behavior of the Variance
The variance of a PERT distribution is commonly approximated by:
$$\operatorname{Var}(X) \approx \left( \frac{b - a}{6} \right)^2 = \frac{(b - a)^2}{36}$$

Given our condition for symmetry where $b = 2c - a$, we can substitute $b$:
$$b - a = (2c - a) - a = 2(c - a)$$

Now, substituting this back into the variance formula:
$$\operatorname{Var}(X) = \frac{[2(c - a)]^2}{36} = \frac{4(c - a)^2}{36}$$

Simplifying the fraction:
$$\operatorname{Var}(X) = \frac{(c - a)^2}{9}$$




When the expected value $E[X]$ is equal to the mode $c$, the distribution becomes **perfectly symmetric**. In this specific case, the mode is exactly halfway between the minimum and maximum values ($11$ is the midpoint of $5$ and $17$).

The variance then depends solely on the squared distance between the most likely time and the boundaries, representing how concentrated the "expert opinion" is around that central estimate.
